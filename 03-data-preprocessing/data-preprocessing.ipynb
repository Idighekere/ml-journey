# Data Preprocessing Tools

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
df=pd.read_csv("Data.csv")
# X=df.drop(columns=["Purchased"]).values
X=df.drop(columns=["Purchased"])
# y=df[["Purchased"]].values
y=df[["Purchased"]]

X
```

## Taking care of missing data

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
imputer.fit(X[["Age", "Salary"]])
X[["Age", "Salary"]] = imputer.transform(X[["Age", "Salary"]])
# imputer.fit(X[:, 1:3])
# X[:, 1:3] = imputer.transform(X[:, 1:3])
X
```

## Encoding categorical data

### Encoding the Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

transformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(transformer.fit_transform(X))
X
```

### Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
y
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print("X_train:",X_train)
print("y_train:",y_train)
print("X_test:",X_test)
print("y_test:",y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train[:,3:5]=sc.fit_transform(X_train[:,3:])
X_test[:,3:5]=sc.fit_transform(X_test[:,3:])
```

```python
print("X_test",X_test)
print("X_train",X_train)
```
